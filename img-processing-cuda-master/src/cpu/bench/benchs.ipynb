```python
import numpy as np
```

```python
f1 = open("nlm.txt", "r")
nlm1 = f1.readlines()
ln = int(len(nlm1)/4)
nlm = np.array([np.array([nlm1[i].split()[0]]).astype(np.int32) for i in range(len(nlm1))])
nlm = np.array([nlm[:ln], nlm[ln:2*ln],  nlm[2*ln:3*ln],  nlm[3 * ln:4*ln]])
nlm = nlm.reshape(4, ln)
f1.close()
```

```python
import matplotlib.pyplot as plt
colors =['ro-','bo-','go-','co-','yo-','ko-','wo-','mo-']
ds = [2,3,4,5]
labels = ['conv_size = ' + str(ds[i]) for i in range(len(ds))]
#plt.yscale('log')
#plt.xscale('log')

plt.xlabel('Radius parameter')
plt.ylabel('Time in s')
#plt.ylim([1e-4, 1e4])
plt.xlim([1, 8])
radius =  [2,3,4,5,6]
for i in range(len(ds)):
    plt.plot(radius, nlm[i,:], colors[i% len(colors)])
plt.legend(labels)
plt.title("Time measurement in non local means for Lena 256 * 256 ")
plt.show()    
```

```python
f1 = open("knn.txt", "r")
knn1 = f1.readlines()
ln = int(len(knn1)/2)
knn = np.array([np.array([knn1[i].split()[0]]).astype(np.int32) for i in range(len(knn1))])
knn = np.array([knn[:ln], knn[ln:2*ln]])
knn = knn.reshape(2, ln)
f1.close()
```

```python
import matplotlib.pyplot as plt
colors =['ro-','bo-','go-','co-','yo-','ko-','wo-','mo-']
ds = [256, 512]
labels = ['Image_size = ' + str(ds[i]) +' * '+str(ds[i]) for i in range(len(ds))]
#plt.yscale('log')
#plt.xscale('log')

plt.xlabel('number of neighbours')
plt.ylabel('Time in s')
#plt.ylim([1e-4, 1e4])
plt.xlim([1, 25])
neigh =  [2 + 2 * i for i in range(12) ]
for i in range(len(ds)):
    plt.plot(neigh, knn[i,:], colors[i% len(colors)])
plt.legend(labels)
plt.title("Time measurement for knn ")
plt.show()    
```
